# Examples of Data Distributions

## Uniform Distribution

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

# Uniformly distributed random set of values between -10 and 10, 100,000 of them.
values = np.random.uniform(-10.0, 10.0, 100000)
plt.hist(values, 50)

# You would see a flat line of the probability distribution function.
plt.show()
```

## Normal / Gaussian

Visualize the probability density function:

```python
from scipy.stats import norm
import matplotlib.pyplot as plt

# Create a list of x values for plotting that range between -3 and 3, increment of
# 0.001.
x = np.arange(-3, 3, 0.001)
# y-axis is the normal function, probability density function for a normal
# distribution on those x values.
plt.plot(x, norm.pdf(x))
```

Generate some random numbers with a normal distribution. "mu" is the desired mean, "sigma" is the standard deviation:

```python
import numpy as np
import matplotlib.pyplot as plt

mu = 5.0
sigma = 2.0
values = np.random.normal(mu, sigma, 10000)
plt.hist(values, 50)
plt.show()
```

## Exponential PDF / "Power Law"

```python
from scipy.stats import expon
import matplotlib.pyplot as plt

# Create values between 0 and 10, step size 0.001.
x = np.arange(0, 10, 0.001)
# Y-axis is the exponential probability distribution function of x.
plt.plot(x, expon.pdf(x))
```

## Binomial Probability Mass Function

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

n, p = 10, 0.5  # Shape parameters
x = np.arange(0, 10, 0.001)
plt.plot(x, binom.pmf(x, n, p))
```

## Poisson Probability Mass Function

The idea here is if you have some information about the average number of things that happen in some given time period, this can give you a way to predict the odds of getting some other value instead on a given future day.

Example: My website gets on average 500 visits per day. What's the odds of getting 550?

```python
from scipy.stats import poisson
import matplotlib.pyplot as plt

mu = 500  # average
x = np.arange(400, 600, 0.5)
plt.plot(x, poisson.pmf(x, mu))
```
